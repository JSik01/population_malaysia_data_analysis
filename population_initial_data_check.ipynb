Checking when age and ethnicity has current values

```python
without_overall_population = population_data[~population_data[["sex", "age", "ethnicity"]]
                                             .isin(["overall", "both"]).any(axis = 1)].copy()
without_overall_population.head()
```

```python
without_overall_population["year"] = pd.to_datetime(without_overall_population["date"]).dt.year
without_overall_population.head()
```

```python
without_overall_population["exists"] = 1
table = without_overall_population.pivot_table(index = "year", columns= "age", values="exists",
                                               aggfunc= "max", fill_value = 0)
table
```

```python
without_overall_population["exists"] = 1
table = without_overall_population.pivot_table(index = "year", columns= "ethnicity", values="exists",
                                               aggfunc= "max", fill_value = 0)
table
```

```python
with_overall_population = population_data[(population_data["age"] == "overall") &
                                          (population_data["ethnicity"] == "overall") &
                                          (population_data["sex"] == "both")].copy()
with_overall_population.head()
```
